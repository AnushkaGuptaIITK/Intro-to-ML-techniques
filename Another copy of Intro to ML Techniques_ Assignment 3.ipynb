```python
import numpy as np
import pandas as pd
import tensorflow
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.utils import resample
from sklearn.ensemble import VotingClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

```python
dataset_path = 'train_weekthree.csv'
df = pd.read_csv(dataset_path)
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)

```

```python
model_nn = Sequential()
model_nn.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
model_nn.add(Dense(1, activation='sigmoid'))
model_nn.compile(loss='binary_crossentropy', optimizer=Adam(), metrics=['accuracy'])
model_nn.fit(X_train, y_train, epochs=10, batch_size=32, verbose=0)
```

```python
model_tree = DecisionTreeClassifier()
model_tree.fit(X_train, y_train)
```

```python
predictions_nn = model_nn.predict(X_test).flatten()
predictions_tree = model_tree.predict(X_test)
predictions_ensemble = np.round((predictions_nn + predictions_tree) / 2)
```

```python
error_nn = 1 - accuracy_score(y_test, np.round(predictions_nn))
error_tree = 1 - accuracy_score(y_test, predictions_tree)
error_ensemble = 1 - accuracy_score(y_test, predictions_ensemble)

accuracy_ensemble = accuracy_score(y_test, np.round(predictions_ensemble))
print(f'Ensemble Accuracy: {accuracy_ensemble}')

print(f'Neural Network Error: {error_nn}')
print(f'Decision Tree Error: {error_tree}')
print(f'Ensemble Error: {error_ensemble}')
```
